```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("ipl_ball_by_ball.csv")

# Remove rows with null batsman or runs_batsman
df = df[df['batsman'].notna() & df['runs_batsman'].notna()]

# Convert runs_batsman to numeric if needed
df['runs_batsman'] = pd.to_numeric(df['runs_batsman'], errors='coerce')

# Now group and plot
top_batsmen = df.groupby('batsman')['runs_batsman'].sum().sort_values(ascending=False).head(10)

if not top_batsmen.empty:
    top_batsmen.plot(kind='bar', title='Top 10 Batsmen (Total Runs)', ylabel='Runs', xlabel='Batsman', color='orange')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No valid data for batsman runs!")

# Most frequent dismissal kinds
sns.countplot(data=df[df['dismissal_kind'].notnull()], x='dismissal_kind')
plt.title('Dismissal Types')
plt.xticks(rotation=45)
plt.show()

# Runs per over
runs_per_over = df.groupby('over')['runs_total'].sum()
sns.lineplot(x=runs_per_over.index, y=runs_per_over.values)
plt.title('Runs per Over (All Matches)')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()

```

```python
df = pd.read_csv("ipl_final_dataset.csv")

# Winner distribution
df['winner'].value_counts().head(10).plot(kind='bar', title='Top 10 Most Wins by Teams', color='skyblue')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Toss vs Match Winner
# Create new column: did toss winner also win?
df['toss_match_win'] = df['toss_winner'] == df['winner']

# Now plot correctly
sns.countplot(x='toss_match_win', data=df)
plt.title('Did Toss Winner Also Win Match?')
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Toss Win = Match Win')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Correlation heatmap (only numerical)
plt.figure(figsize=(14, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm', annot=True, fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Ensure your date column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the year from the date
df['year'] = df['date'].dt.year

# Filter for years 2008 to 2025
df_filtered = df[df['year'].between(2008, 2025)]

# Set up the FacetGrid
g = sns.FacetGrid(df_filtered, col='year', col_wrap=5, height=3.5, sharex=True, sharey=False)

# Draw the countplot on each facet
g.map(sns.countplot, 'toss_winner', order=[0, 1])

# Add titles and labels
g.set_titles(col_template="Year {col_name}")
g.set_axis_labels("Toss Win = Match Win", "Count")

# Customize the x-tick labels
for ax in g.axes.flatten():
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['No', 'Yes'])

plt.subplots_adjust(top=0.92)
g.fig.suptitle("Did Toss Winner Also Win Match? (Year-wise)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ipl_match_by_match.csv")

# Fill NA and categorize
df['result'] = df['result'].fillna('no result')

# Categorize
def simplify_result(result):
    if 'wickets' in result:
        return 'Won by Wickets'
    elif 'runs' in result:
        return 'Won by Runs'
    elif result == 'no result':
        return 'No Result'
    else:
        return 'Other'

df['result_category'] = df['result'].apply(simplify_result)

# Plot pie chart with grouped categories
df['result_category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05]*3,
    colors=['skyblue', 'lightgreen', 'lightcoral'],
    wedgeprops={'linewidth': 1, 'edgecolor': 'black'}
)
plt.title('Result Type Distribution (Grouped)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
file_path = '/ipl_match_by_match.csv'
df = pd.read_csv(file_path)

sns.set(style='whitegrid')

# 1. Overall Winning Stats
print("Overall Winning Stats (% wins per team):")
overall_wins = df['winner'].value_counts(normalize=True) * 100
print(overall_wins.round(2))

# 2. Winning Team Based on Venue (normalized)
print("\nWinning Team Based on Venue (% per venue):")
venue_team_wins = (
    df.groupby(['venue', 'winner']).size().unstack(fill_value=0)
    .div(df.groupby('venue')['winner'].count(), axis=0) * 100
)
print(venue_team_wins.round(2))

# 3. Pitch Friendliness: Batsmen vs Bowlers
print("\nPitch Friendliness (if toss-winner wins, % by toss decision):")
batsmen_mask = (df['toss_winner'] == df['winner']) & (df['toss_decision'] == 'bat')
bowlers_mask = (df['toss_winner'] == df['winner']) & (df['toss_decision'] == 'field')
bat_pct = batsmen_mask.sum() / len(df) * 100
bowl_pct = bowlers_mask.sum() / len(df) * 100
print(f"Batsmen Win %: {bat_pct:.2f}")
print(f"Bowlers Win %: {bowl_pct:.2f}")

# 4. Chasing vs Defending (per team, normalized)
print("\nChasing Wins (% by team):")
chasing_wins = (
    df[df['team1_score'] < df['team2_score']]['winner'].value_counts(normalize=True) * 100
)
print(chasing_wins.round(2))

print("\nDefending Wins (% by team):")
defending_wins = (
    df[df['team1_score'] > df['team2_score']]['winner'].value_counts(normalize=True) * 100
)
print(defending_wins.round(2))

# 5. Best Team at Chasing
print("\nBest Team at Chasing:")
best_chasing_team = (
    df[df['team1_score'] < df['team2_score']]['winner'].value_counts().idxmax()
)
print(best_chasing_team)

# 6. Performance at Specific Stadiums (≥5 matches)
print("\nPerformance at Venues With ≥5 Matches (% by team):")
venues_5plus = df['venue'].value_counts()
selected_venues = venues_5plus[venues_5plus >= 5].index
venue_perf = (
    df[df['venue'].isin(selected_venues)]
    .groupby(['venue', 'winner']).size().unstack(fill_value=0)
    .div(df[df['venue'].isin(selected_venues)].groupby('venue')['winner'].count(), axis=0) * 100
)
print(venue_perf.round(2))

# === OPTIONAL: Visualization Section ===
# (Uncomment if you want quick plots as well)

# Pie chart for Overall Winning Stats
plt.figure(figsize=(7,7))
overall_wins.plot.pie(autopct='%1.1f%%')
plt.title('Overall Winning Stats')
plt.ylabel('')
plt.show()

# Heatmap for Winning Team Based on Venue
plt.figure(figsize=(len(venue_team_wins.columns)*1.5, len(venue_team_wins)*0.6))
sns.heatmap(venue_team_wins, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Team Win % per Venue')
plt.show()

# Bar chart for Pitch Friendliness
plt.bar(['Batsmen', 'Bowlers'], [bat_pct, bowl_pct], color=['blue', 'red'])
plt.ylabel('Win Percentage')
plt.title('Pitch Friendliness')
plt.show()
```
